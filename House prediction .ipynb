```python
# Assuming df is your DataFrame
def is_float(x):
    try:
        float(x)
        return True
    except:
        return False
```

```python
# Checking which entries are not directly convertible to float values
non_float_entries = df3[~df3['total_sqft'].apply(is_float)]['total_sqft']
print("Entries that are not directly convertible to float values:")
print(non_float_entries)
```

```python
# Converting the 'total_sqft' column to float where possible
df3['total_sqft'] = df3[df3['total_sqft'].apply(is_float)]['total_sqft'].astype(float)
```

```python
def convert_sqft_to_num(x):
    if '-' in str(x):
        tokens = str(x).split('-')
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1])) / 2
    try:
        return float(x)
    except:
        return None

df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4 = df4[df4['total_sqft'].notnull()]
print(df4.head(5))
```

# Feature Engineering Adding a new feature 'price_per_sqft'

```python

# Apply the conversion function to the 'total_sqft' column
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)

# Add a new feature 'price_per_sqft'
df4['price_per_sqft'] = df4['price']*100000 / df4['total_sqft']

# Display the updated DataFrame
print(df4)
```

```python
#Here we find that min price per sqft is 267 rs/sqft whereas max is 12000000, 
#this shows a wide variation in property prices. We should remove outliers per location using mean and one standard deviation
import pandas as pd
import numpy as np

def remove_pps_outliers(df):
    return df.groupby('location').apply(lambda x: x[(x['price_per_sqft'] > (x['price_per_sqft'].mean() - x['price_per_sqft'].std())) & 
                                                   (x['price_per_sqft'] <= (x['price_per_sqft'].mean() + x['price_per_sqft'].std()))])

df6 = remove_pps_outliers(df5)
print(df6.shape)
```

```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            if bhk > 1:
                prev_stats = bhk_stats.get(bhk - 1)
                if prev_stats:
                    exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < prev_stats['mean']].index.values)
    return df.drop(exclude_indices, axis='index')

df8 = remove_bhk_outliers(df7)
print(df8.shape)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

# Assuming 'X' contains the features and 'Y' contains the target variable
regressor = LinearRegression()

# Perform 5-fold cross-validation
scores = cross_val_score(regressor, X, Y, cv=5, scoring='neg_mean_squared_error')

# Print the accuracy scores
print('Accuracy:', -scores)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

# Create a list of dictionaries containing the models and their respective parameters
models_and_parameters = [
    (LinearRegression(), {}),
    (RandomForestRegressor(), {'n_estimators': [10, 50, 100]}),
    (DecisionTreeRegressor(), {'max_depth': [None, 5, 10]})
]

# Create a GridSearchCV object
grid_searches = []
for model, params in models_and_parameters:
    grid_search = GridSearchCV(estimator=model, param_grid=params, scoring='neg_mean_squared_error', cv=5)
    grid_searches.append(grid_search)

# Fit the GridSearchCV objects to the data
for grid_search in grid_searches:
    grid_search.fit(X, Y)

# Get the best parameters and best scores for each model
for i, grid_search in enumerate(grid_searches):
    best_params = grid_search.best_params_
    best_score = grid_search.best_score_
    print(f"Best Parameters for model {i+1}: {best_params}")
    print(f"Best Score for model {i+1}: {best_score}")
```

```python
def predict_price(location, sqft, bath, bhk, model, columns):
    input_data = np.zeros(len(columns))
    input_data[0] = sqft
    input_data[1] = bath
    input_data[2] = bhk
    if location in columns:
        loc_index = columns.index(location)
        input_data[loc_index] = 1

    return model.predict([input_data])[0]
```
